```
from datetime import timedelta, datetime, date

table_name = 'deep_sentinel.event_partition2'
create_body = '  `uuid` varchar(40) NOT NULL,\n  `date_time` datetime(3) DEFAULT NULL'
partition_field = 'date_time'
start_date = date(2019, 12, 31)
day_count = 750
engine = 'InnoDB'

print ('CREATE TABLE ' + table_name + ' (')
print (create_body)
print (') PARTITION BY RANGE(TO_DAYS(' + partition_field + '))')
print ('(')
last_date = start_date + timedelta(day_count-1)
for single_date in (start_date + timedelta(n) for n in range(day_count)):
  close_string = ','
  if single_date == last_date:
    close_string = ''
  print ('  partition ' + single_date.strftime("p%Y_%m_%d")  + ' VALUES LESS THAN ( TO_DAYS(\'' + single_date.strftime("%Y-%m-%d") + '\')) ENGINE = ' + engine + close_string)
print (');')  
```
