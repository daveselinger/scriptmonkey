```
from datetime import timedelta, datetime, date

# A script that will move data from one table to another (exact same schema) one day at a time.
# this is needed if the source table is too big to do a single transfer and would overflow buffers, cause full table scans etc.

# The name of the table used for input
source_table_name = 'event_log'
source_date_column_name = 'date_time'

# The name of the table used for output
dest_table_name = 'event_log_partition'

# The date range will be essentially -- one partition which will catch all events before start_date
# Followed by "day_count" partitions each with 1 day of data
start_date = date(2019, 1, 1)
day_count = 1100

engine = 'InnoDB'

last_date = start_date + timedelta(day_count-1)
one_day = timedelta(1)

for single_date in (start_date + timedelta(n) for n in range(day_count)):
  print ('INSERT INTO ' + dest_table_name + ' SELECT * from ' + source_table_name + ' where '+ source_date_column_name + ' >= \'' +  (single_date).strftime("%Y-%m-%d") + '\' AND date_time < \'' +  (single_date + one_day).strftime("%Y-%m-%d") +'\'')
```
